```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
import random
```

```python
df = pd.read_excel('Household Income.xls')
df
```

```python
df_flipped = df.transpose().reset_index()
df_flipped.columns = df_flipped.iloc[0]
df_flipped = df_flipped.rename({'States': 'Years'}, axis=1)
df_flipped = df_flipped[1:].iloc[::-1].reset_index().drop(['index'], axis = 1)
#df_flipped['Years'] = df_flipped['Years'].str.rstrip('egatnecreP_').astype({'Years': 'int'})
df_flipped
```

```python
plt.figure(figsize=(24,13))
plt.plot( 'Years', 'United States', data=df_flipped, marker='*', color='#4832a8', linewidth=5)
plt.plot( 'Years', 'Illinois', data=df_flipped, marker='*', markersize=2, color='#3dff47', linewidth=1)
plt.plot( 'Years', 'Alaska', data=df_flipped, marker='*', markersize=2, color='olive', linewidth=1)
plt.plot( 'Years', 'Arizona', data=df_flipped, marker='p', markersize=2, color='skyblue', linewidth=1)
plt.plot( 'Years', 'Arkansas', data=df_flipped, marker='p', markersize=2, color='#32a2a8', linewidth=1)
plt.plot( 'Years', 'California', data=df_flipped, marker='p', markersize=2, color='#a83232', linewidth=1)
plt.plot( 'Years', 'Texas', data=df_flipped, marker='p', markersize=2, color='#ffbb3d', linewidth=1)
plt.legend()
plt.xlabel('Years', fontsize = 'xx-large')
plt.ylabel('Percentage of Uninsured People', fontsize = 'xx-large')
```

```python
plt.figure(figsize=(24,13))
plt.plot( 'Years', 'United States', data=df_flipped, marker='*', color='#4832a8', linewidth=5)
for each_state in df_flipped.columns[2:]:
    plt.plot( 'Years', each_state, data=df_flipped, marker='p', markersize=2, color=(random.random(), random.random(), random.random()), linewidth=1)
plt.legend()
plt.xlabel('Years', fontsize = 'xx-large')
plt.ylabel('Percentage of Uninsured People', fontsize = 'xx-large')
```
