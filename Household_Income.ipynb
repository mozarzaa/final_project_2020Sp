```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
import random
```

```python
df = pd.read_excel('Household Income.xls')
df
```

```python
df_flipped = df.transpose().reset_index()
df_flipped.columns = df_flipped.iloc[0]
df_flipped = df_flipped.rename({'States': 'Years'}, axis=1)
df_flipped = df_flipped[1:].iloc[::-1].reset_index().drop(['index'], axis = 1)
#df_flipped['Years'] = df_flipped['Years'].str.rstrip('egatnecreP_').astype({'Years': 'int'})
df_flipped
```

```python
plt.figure(figsize=(24,13))
plt.plot( 'Years', 'United States', data=df_flipped, marker='*', color='#4832a8', linewidth=5)
plt.plot( 'Years', 'Illinois', data=df_flipped, marker='*', markersize=2, color='#3dff47', linewidth=1)
plt.plot( 'Years', 'Alaska', data=df_flipped, marker='*', markersize=2, color='olive', linewidth=1)
plt.plot( 'Years', 'Arizona', data=df_flipped, marker='p', markersize=2, color='skyblue', linewidth=1)
plt.plot( 'Years', 'Arkansas', data=df_flipped, marker='p', markersize=2, color='#32a2a8', linewidth=1)
plt.plot( 'Years', 'California', data=df_flipped, marker='p', markersize=2, color='#a83232', linewidth=1)
plt.plot( 'Years', 'Texas', data=df_flipped, marker='p', markersize=2, color='#ffbb3d', linewidth=1)
plt.legend()
plt.xlabel('Years', fontsize = 'xx-large')
plt.ylabel('Household Income (USD)', fontsize = 'xx-large')
```

```python
plt.figure(figsize=(24,13))
plt.plot( 'Years', 'United States', data=df_flipped, marker='*', color='#4832a8', linewidth=5)
for each_state in df_flipped.columns[2:]:
    plt.plot( 'Years', each_state, data=df_flipped, marker='p', markersize=2, color=(random.random(), random.random(), random.random()), linewidth=1)
plt.legend()
plt.xlabel('Years', fontsize = 'xx-large')
plt.ylabel('Household Income (USD)', fontsize = 'xx-large')
```

```python
inc_exp_2013_2018 = pd.read_excel('multiyr_2013_2018.xlsx',skiprows=[0,1],nrows=219)
inc_exp_2013_2018=inc_exp_2013_2018[inc_exp_2013_2018['Item'].isna()==False]
row_list=['Income before taxes',
          'Average annual expenditures',
          'Health insurance**',
          'Life and other personal insurance',]
inc_exp_2013_2018=inc_exp_2013_2018[inc_exp_2013_2018['Item'].isin(row_list)]
inc_exp_2013_2018=inc_exp_2013_2018.reset_index().drop(['index'], axis = 1)
inc_exp_2013_2018.loc[2, 'Item']='Health insurance'

inc_exp_2013_2018
```

```python
inc_exp_2006_2012 = pd.read_excel('multiyr_2006_2012.xls',skiprows=[0,1],nrows=217)
inc_exp_2006_2012=inc_exp_2006_2012[inc_exp_2006_2012['Item'].isna()==False]
row_list=['Income before taxes a/',
          'Average annual expenditures',
          'Health insurance',
          'Life and other personal insurance',]
inc_exp_2006_2012=inc_exp_2006_2012[inc_exp_2006_2012['Item'].isin(row_list)]
inc_exp_2006_2012=inc_exp_2006_2012.reset_index().drop(['index'], axis = 1)
inc_exp_2006_2012.loc[0, 'Item']='Income before taxes'
#inc_exp_2006_2012=inc_exp_2006_2012.drop(['Income after taxes a/'])

inc_exp_2006_2012
```

```python
inc_exp_2000_2005 = pd.read_excel('multiyr_2000_2005.xls',skiprows=[0,1],nrows=170)
inc_exp_2000_2005=inc_exp_2000_2005.drop(['Unnamed: 7','Unnamed: 8'], axis = 1)
row_list=['Income before taxes 1/ ',
          'Average annual expenditures',
          '  Health insurance',
          '  Life and other personal insurance',]
inc_exp_2000_2005.columns=['Item','2000','2001','2002','2003','2004','2005']
inc_exp_2000_2005=inc_exp_2000_2005[inc_exp_2000_2005['Item'].isin(row_list)]
inc_exp_2000_2005=inc_exp_2000_2005.reset_index().drop(['index'], axis = 1)
inc_exp_2000_2005.loc[0, 'Item']='Income before taxes'
inc_exp_2000_2005.loc[2, 'Item']='Health insurance'
inc_exp_2000_2005.loc[3, 'Item']='Life and other personal insurance'

inc_exp_2000_2005
```

```python
inc_exp_merge=pd.merge(inc_exp_2006_2012,inc_exp_2013_2018, how='inner', on='Item')
inc_exp_merge=pd.merge(inc_exp_2000_2005,inc_exp_merge, how='inner', on='Item')

inc_exp_merge
```

```python
inc_exp_flipped = inc_exp_merge.transpose().reset_index()
inc_exp_flipped.columns = inc_exp_flipped.iloc[0]
inc_exp_flipped = inc_exp_flipped.rename({'Item': 'Years'}, axis=1)
inc_exp_flipped = inc_exp_flipped[1:].iloc[::-1].reset_index().drop(['index'], axis = 1)
#df_flipped['Years'] = df_flipped['Years'].str.rstrip('egatnecreP_').astype({'Years': 'int'})
inc_exp_flipped
```

```python
plt.figure(figsize=(24,13))
plt.plot('Years', 'Health insurance', data=inc_exp_flipped, marker='*', color='#4832a8', linewidth=5)
for each_state in inc_exp_flipped.columns[2:]:
    plt.plot( 'Years', each_state, data=inc_exp_flipped, marker='p', markersize=2, color=(random.random(), random.random(), random.random()), linewidth=1)
plt.legend()
plt.xlabel('Years', fontsize = 'xx-large')
plt.ylabel('Household Income and Insurance Spending (USD)', fontsize = 'xx-large')
```
